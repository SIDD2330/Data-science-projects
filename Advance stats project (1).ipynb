Ho = μBefore−μAfter = 5 (the training will not make a difference of more than 5)

Ha = μBefore−μAfter < 5 (the training will  make a difference of less than 5)

H0:  μ(pushups) <= 5

This is a problem of the paired-t-test (level of Significance as 5%)

```python
def remove_outlier(col):
    sorted(col)
    Q1,Q3=col.quantile([0.25,0.75])
    IQR=Q3-Q1
    lower_range= Q1-(1.5 * IQR)
    upper_range= Q3+(1.5 * IQR)
    return lower_range, upper_range
```

```python
lrResponse,urResponse=remove_outlier(df2['Response'])
df2['Response']=np.where(df2['Response']>urResponse,urResponse,df2['Response'])
df2['Response']=np.where(df2['Response']<lrResponse,lrResponse,df2['Response'])
```
